Standardization of Numerical Variables

Objective: To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.

Explain the concept of standardization (z-score normalization).

* Z score normalization is a data preprocessing technique using in Machine Learning and Statsitcs.
* Z score normalization transforms the data so that it has the mean of 0 and standard deviation of 1.
* It is used to compare values of normally distributed datasets.
* Z score is a measure of how many standard deviation a data point is from its population mean.

Interpreting Z Score:

* z score typiclly range from -3 to +3.
* So if, 
    * z score=0, the value is equal to the mean.
    * Postitive z score, the value is greater than the mean.
    * Negative z score, the value is less than the mean.

                                                  z score Formula:
$$ Z =  \frac{X - \mu}{\sigma} $$

Where:

- **X** = Raw score (data value)
- **\(Î¼\)** = Population mean
- **\(Ïƒ\)** = Population standard deviation

So if z score results in 2.2, that means the selected data point is 2.2 standard deviation above the mean.
And if z score results in .1.5, that means the selected data point is 1.5 standard deviation below the mean.

```python
numerical_columns=['Volume','Avg Price','Total Sales Value','Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
```

```python
for i in numerical_columns:
    print(f"Mean of {i} is = {df[i].mean()} \nStantard Deviation of {i} is = {df[i].std()}\n")
```

```python
for i in numerical_columns:
    print(f"Mean of {i} is = {round(df[i].mean())} \nStantard Deviation of {i} is = {round(df[i].std())}\n")
```

Discuss the need for converting categorical data into dummy variables (one-hot encoding).  
One Hot Encoding is a process of converting categorical variables to numerical to be specific into binary values its either 0 or 1.  
One hot encoding with this conversion allows each category in the categorical variables to be treated indepentently thus it does not imply any false relationship between them.  
And also many Machine Learning Algorithms which works primarily on numerical columns, the categorical columns needs to be transformed for them to work in a much effective way.  
Also, one hot encoding's concept of treating each category indepentently eleminates the ordinality as label encoding which assigns the rank to these category will cause a problem by suggesting an order or relationship which typically does not exists.

#### **Importance of Data Preprocessing (Standardization & One-Hot Encoding)**  
1. **Standardization (Z-Score Normalization)**  
   - Standardizing numerical variables ensured that all data had a mean of **0** and a standard deviation of **1**, preventing features with large values (e.g., sales value) from dominating smaller ones (e.g., discount rate).  
   - This step was crucial for improving model stability and convergence in machine learning applications.  

2. **One-Hot Encoding (OHE) for Categorical Variables**  
   - Categorical columns such as **SKU, Brand, Model, BU, and Day** were transformed into binary (0 or 1) format.  
   - This prevented the algorithm from assuming any ordinal relationship between categories, unlike label encoding.  
   - Encoding was necessary for models that do not handle categorical variables directly, ensuring proper feature representation.  

By implementing these preprocessing steps, we ensured that the dataset was well-structured for further predictive modeling and analytics.
