```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove extreme outliers
# for i in df.columns:
#     if i not in ["DS", "DP", "NSP"]:  # Corrected condition
#         df[i] = df[i].clip(lower=lower_bound[i], upper=upper_bound[i])

# Remove rows with outliers (excluding DS, DP, NSP)
for col in df.columns:
    if col not in ["DS", "DP", "NSP"]:
      df = df[(df[col] >= lower_bound[col]) & (df[col] <= upper_bound[col])]
```

### **Interpretation of Skewness & Kurtosis Results**  

#### **1. Skewness Interpretation**

  - **LB (-0.18)**, **UC (-0.09)**, **MLTV (0.18)**, and **ASTV (0.14)** are approximately **symmetrically distributed**.  

- **Moderate Positive Skewness (Right-Skewed)**:

- **Highly Positive Skewness (Strong Right-Skewed)**:  
  - **FM (1.16)**, **DL (1.26)**, and **ALTV (1.31)** show **strong right-skew**, meaning most values are concentrated on the left with some extreme high values.  

- **Missing Skewness (NaN values)**:  
  - **DS, DP, NSP** have NaN skewness, likely because they contain only a single unique value (constant data).  

---

#### **2. Kurtosis Interpretation**  
- **Negative Kurtosis (< 3, Platykurtic - Light tails, fewer outliers)**:  
  - **Most variables have negative kurtosis**, meaning they have **flatter distributions** with **lighter tails** (fewer extreme values).  
  - Examples: **AC (-0.54)**, **FM (-0.64)**, **UC (-0.56)**, **ASTV (-1.02)**, **MSTV (-0.02)**, **MLTV (-0.35)**, **Width (-0.31)**, **Tendency (-1.82)**.  

- **Kurtosis near Zero (Close to Normal Distribution)**:  
  - **LB (-0.26)**, **DL (0.33)**, and **ALTV (0.03)** are **close to a normal distribution**.  

- **Missing Kurtosis (NaN values)**:  
  - **DS, DP, NSP** have NaN kurtosis, likely due to having constant values.  

---

### **Key Insights**  
* **Highly skewed variables (FM, DL, ALTV) may require transformation (log, sqrt, or Box-Cox transformation) to make them more normal.**  
* **Most features have negative kurtosis, indicating they are flatter than a normal distribution (fewer extreme values).**
