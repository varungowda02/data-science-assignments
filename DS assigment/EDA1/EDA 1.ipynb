```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in df_new.columns:
    plt.figure(figsize=(6, 3))
    sns.histplot(df_new[i], bins=30, kde=True)
    plt.title(f"Distribution of {i}")
    plt.show()
```

```python
# Compute Q1, Q3, and IQR for each column
Q1 = df_new.quantile(0.25)
Q3 = df_new.quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count outliers in each column
outliers = ((df_new < lower_bound) | (df_new > upper_bound)).sum()
print("Number of outliers per column:\n", outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 10))

# Before treatment
plt.subplot(1, 2, 1)
sns.boxplot(data=df_new)
plt.xticks(rotation=90)
plt.title("Before Outlier Treatment")
```

```python
# Create a copy to avoid modifying the original data
df2 = df_new.copy()

# Replace outliers with median
for col in df_new.columns:
    median_value = df_new[col].median()
    df2[col] = np.where(
        (df_new[col] < lower_bound[col]) | (df_new[col] > upper_bound[col]),
        median_value,  # Replace outliers with median
        df_new[col]    # Keep original value if not an outlier
    )
```

```python
# Check if outliers remain
outliers_after = ((df2 < lower_bound) | (df2 > upper_bound)).sum()
print("Number of outliers per column after replacement:\n", outliers_after)
```

###3. Data Visualization:
* Create histograms or boxplots to visualize the distributions of various numerical variables.
* Use bar charts or pie charts to display the frequency of categories for categorical variables.
* Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.
* Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style="whitegrid")

# Create histograms for all numerical columns
for col in df.columns:
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col], bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Bar charts (by binning numerical values)
for col in df_new.columns:
    plt.figure(figsize=(6, 4))
    df_new[col].plot(kind='hist', bins=10, color="goldenrod", alpha=0.7, edgecolor="black")
    plt.title(f"Bar Chart of {col} (Binned)")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter Plots: Pairwise relationships
for col1 in df_new.columns:
    for col2 in df_new.columns:
        if col1 != col2:  # Avoid self-correlation plots
            plt.figure(figsize=(6, 4))
            sns.scatterplot(x=df_new[col1], y=df_new[col2], alpha=0.5, color="royalblue")
            plt.title(f"Scatter Plot: {col1} vs {col2}")
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.show()

```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_new.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pair Plot - Visualizing relationships between all numerical variables
sns.pairplot(df_new, diag_kind="kde")
plt.show()


```

```python
# Violin Plots - Distribution of each numerical variable
plt.figure(figsize=(15, 8))
for i, col in enumerate(df_new.columns, 1):
    plt.subplot(3, 5, i)  # Adjust layout for better display
    sns.violinplot(y=df_new[col], color="skyblue")
    plt.title(f"Violin Plot of {col}")
    plt.xlabel("")
plt.tight_layout()
plt.show()
```

###4. Pattern Recognition and Insights:
* Identify any correlations between variables and discuss their potential implications.
* Look for trends or patterns over time if temporal data is available.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df_new.corr()

# Heatmap for Correlation Analysis
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Scatter Plots for Strongly Correlated Pairs (Threshold: |correlation| > 0.7)
strong_corr = corr_matrix.abs().unstack().sort_values(ascending=False)
strong_corr = strong_corr[strong_corr.between(0.7, 1) & (strong_corr < 1)]

if not strong_corr.empty:
    for (col1, col2), value in strong_corr.items():
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=df_new[col1], y=df_new[col2], alpha=0.6)
        plt.title(f"Scatter Plot: {col1} vs {col2} (Corr: {value:.2f})")
        plt.xlabel(col1)
        plt.ylabel(col2)
        plt.show()
else:
    print("No strongly correlated variable pairs found.")

# Check Trends Over Time if Temporal Data Exists
if "Time" in df_new.columns:  # Replace 'Time' with the actual time column if available
    plt.figure(figsize=(12, 5))
    sns.lineplot(x=df_new["Time"], y=df_new["LB"])  # Example using "LB"
    plt.title("Trend of LB Over Time")
    plt.xlabel("Time")
    plt.ylabel("LB")
    plt.show()
else:
    print("No temporal data found to analyze trends.")
```
