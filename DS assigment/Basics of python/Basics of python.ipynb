Exercise 1: Prime Numbers

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
var_number = input("Enter a number to check weather its a prime number or not")
# Get input from the user and store it in the variable var_number.
var_number = int(var_number)
# Convert the input string to an integer.

# Check if the number is less than or equal to 1.
# If it is, print an error message and exit.
if(var_number<=1):
    print("Enter a valid number as Prime Number can not be less that 2")

# If the number is greater than 1, proceed with the prime number check.
else:
    # Assume the number is prime initially.
    is_Prime = True
    # Iterate through numbers from 2 to var_number - 1.
    for i in range(2,var_number):
        # If var_number is divisible by i, it's not prime.
        if(var_number % i == 0):
            # Set is_Prime to False and break the loop.
            is_Prime = False
            break;

    # If is_Prime is still True after the loop, the number is prime.
    if is_Prime:
        print(f"{var_number} is a Prime number")
    # Otherwise, it's not prime.
    else:
        print(f"{var_number} is not a prime number")
```

Exercise 3: Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
print("Select\n1: Square of Even numbers from 100 to 200\n2: Square of Odd numbers from 100 to 200")
# Prints a menu for the user to select either even or odd numbers.
snum = input()
# Takes the user's input (1 for even, 2 for odd).
snum = int(snum)
# Converts the user's input to an integer.

if(snum == 1):
    # If the user selected 1 (even numbers):
    for i in range(100,201):
        # Iterates through numbers from 100 to 200 (inclusive).
        if(i % 2 == 0):
            # If the current number (i) is even (divisible by 2):
            print(f"{i} * {i} = {pow(i,2)}")
            # Prints the square of the even number.

elif(snum == 2):
    # If the user selected 2 (odd numbers):
    for i in range(100,201):
        # Iterates through numbers from 100 to 200 (inclusive).
        if(i % 2!= 0):
            # If the current number (i) is odd (not divisible by 2):
            print(f"{i} * {i} = {pow(i,2)}")
            # Prints the square of the odd number.
```

Exercise 4: Word counter

Write a program to count the number of words in a given text.

Example:

Input_text = "This is a sample text. This text will be used to demonstrate the word counter."

Expected output:
'This': 2
'is': 1
'a': 1
'sample': 1
'text.': 1


```python
import re
# Imports the 're' module for regular expression operations.
word_str = input("Enter a sentence")
# Prompts the user to enter a sentence and stores it in 'word_str'.
word_str = re.sub(r"[^\w\s]","", word_str)
# Removes punctuation from the sentence using a regular expression.
words = word_str.split()
# Splits the sentence into a list of words.
word_dict = {}
# Initializes an empty dictionary to store word counts.

for i in words:
    # Iterates through each word in the list of words.
    if i in word_dict:
        # If the word is already in the dictionary (as a key):
        word_dict[i] = word_dict[i] + 1
        # Increments the count of that word by 1.
    else:
        # If the word is not in the dictionary:
        word_dict[i] = 1
        # Adds the word to the dictionary with a count of 1.

print("\n")
# Prints a newline for formatting.
for j, c in word_dict.items():
    # Iterates through the key-value pairs (word and count) in the dictionary.
    print(f"'{j}': {c}")
    # Prints the word and its count in the desired format.

#We all make rules for ourselves. Itâ€™s these rules that help define who we are. So, when we break those rules, we risk losing ourselves and becoming something unknown.
# This is a sample sentence provided as a comment.
```

Exercise 5: Check for Palindrome

Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.

Example:
Input: "racecar"
Expected Output: True

```python
import re
# Imports the 're' module for regular expression operations.
wrd = input("Enter the string to check for Palindrom")
# Prompts the user to enter a string and stores it in 'wrd'.
wrd = wrd.lower().strip()
# Converts the string to lowercase and removes leading/trailing spaces.
wrd = re.sub(r"[^\w]", "", wrd)
# Removes any non-alphanumeric characters from the string.


def is_palindrom(word):
    # Defines a function called 'is_palindrom' that takes a 'word' as input.
    if(word == word[::-1]):
        # Checks if the word is the same when reversed (using slicing).
        print(f"{word} is indeed a Palindrom")
        # If it is, prints a message indicating it's a palindrome.
    else:
        print(f"{word} is NOT a Palindrom")
        # If it's not, prints a message indicating it's not a palindrome.

is_palindrom(wrd)
# Calls the 'is_palindrom' function with the processed string 'wrd'.

#A man, a plan, a canal: Panama!
# This is a sample palindrome provided as a comment.
```
