```python
from mlxtend.preprocessing import TransactionEncoder
encoder=TransactionEncoder()
encoder_a=encoder.fit(transactions).transform(transactions)
```

```python
df=pd.DataFrame(encoder_a, columns=encoder.columns_)
df.head()
```

```python
for i in df.columns:
  print(df[i].value_counts(), "\n")
```

```python
from mlxtend.frequent_patterns import apriori
frequent_items=apriori(df,min_support=0.01, use_colnames=True)
```

```python
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_items, metric='confidence', min_threshold=0.5)
```

```python
rules
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head()
```

```python
rules=rules[rules['lift']>1].sort_values(by='lift', ascending=False)
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head()
```

```python
for i, r in rules.head(10).iterrows():
  antecedents=list(r['antecedents'])
  consequents=list(r['consequents'])
  print(f"Rule: {antecedents} --> {consequents}")
  print(f"Support: {r['support']:.4f}")
  print(f"Confidence: {r['confidence']:.4f}")
  print(f"Lift: {r['lift']:.4f}\n")
```

# **Interview Questions**

## **1. What is Lift, and why is it important?**
- Lift measures the **strength of association** between items.
- **Formula:**  
  $$
  \text{Lift} = \frac{\text{Confidence}}{\text{Expected Confidence (if independent)}}
  $$
- Interpretation:
  - **Lift > 1** â†’ Items occur together **more than expected** (strong association).
  - **Lift = 1** â†’ Items occur **independently**.
  - **Lift < 1** â†’ Items are **negatively associated**.

## **2. What are Support and Confidence?**
- **Support**: How frequently an itemset appears in transactions.
  $$
  \text{Support} = \frac{\text{Transactions containing item A and B}}{\text{Total Transactions}}
  $$
- **Confidence**: How often **B is bought when A is bought**.
  $$
  \text{Confidence} = \frac{\text{Support(A âˆª B)}}{\text{Support(A)}}
  $$

## **3. What are some limitations or challenges of Association rules mining?**
- Requires **large datasets** for meaningful insights.
- Choosing the **right support and confidence thresholds** is tricky.
- Can generate **too many rules**, making it hard to interpret results.
