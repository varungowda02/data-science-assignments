### **Interpretation of Skewness and Kurtosis for Each Feature**  

#### **1. PassengerId**  
- **Skewness: 0.0** → Perfectly symmetrical distribution.  
- **Kurtosis: -1.20** → Platykurtic distribution (flatter than normal).  
- **Interpretation**: This is an ID variable, and its distribution should ideally be uniform and not affect the model.  

#### **2. Survived**  
- **Skewness: 0.48** → Slightly right-skewed (more 0s than 1s).  
- **Kurtosis: -1.77** → Platykurtic (less extreme values than a normal distribution).  
- **Interpretation**: The survival variable has a nearly balanced distribution but leans slightly toward non-survivors.  

#### **3. Pclass**  
- **Skewness: -0.63** → Moderately left-skewed (more passengers in higher classes).  
- **Kurtosis: -1.28** → Platykurtic (flatter than normal).  
- **Interpretation**: The majority of passengers belong to the lower-class categories (e.g., 3rd class), with fewer in the upper-class categories.  

#### **4. Age**  
- **Skewness: NaN, Kurtosis: NaN** → Likely due to missing values.  
- **Interpretation**: Age has missing data, which should be handled via imputation before analysis.  

#### **5. SibSp (Number of Siblings/Spouses Aboard)**  
- **Skewness: 3.69** → Highly right-skewed (most passengers had few or no siblings/spouses).  
- **Kurtosis: 17.77** → Highly leptokurtic (heavy-tailed distribution with many outliers).  
- **Interpretation**: Most passengers traveled alone or with one family member, but a few had large families, leading to extreme values.  

#### **6. Parch (Number of Parents/Children Aboard)**  
- **Skewness: 2.74** → Highly right-skewed (most passengers had no parents/children aboard).  
- **Kurtosis: 9.72** → Leptokurtic (many extreme values).  
- **Interpretation**: Most passengers traveled alone, but some had larger families, creating a heavy-tailed distribution.  

#### **7. Fare**  
- **Skewness: 4.78** → Extremely right-skewed (a few passengers paid very high fares).  
- **Kurtosis: 33.20** → Extremely leptokurtic (very heavy-tailed).  
- **Interpretation**: Most fares were low, but a few first-class passengers paid very high amounts, causing extreme outliers.  

---

### **Overall Insights and Recommendations**  
1. **Highly skewed features (SibSp, Parch, Fare) should be transformed** to reduce skewness (e.g., using log transformation or Box-Cox transformation).  
2. **Age requires handling missing values** before further analysis.  
3. **Fare has extreme outliers**, so consider capping extreme values to improve model robustness.  
4. **Pclass and Survived distributions seem reasonable**, but class imbalance should be checked.  
5. **PassengerId is not useful for analysis** and should be excluded from model training.

The shape of a DataFrame represents its dimensions: the number of rows and columns. In my case, the original shape was **891 rows × 14 columns** which is larger than **(749, 14)**. Removing outliers using the Z-score method resulted in the removal of rows (data points) that were considered outliers based on their Z-scores. This is why the number of rows decreased to 749.   

Z-score is a statistical measure that quantifies how many standard deviations a data point is away from the mean. It's often used to identify and remove outliers, which are data points significantly different from the rest of the data. Typically, data points with a Z-score above 3 or below -3 are considered outliers and removed.  

**Impact on Accuracy**

Removing outliers can sometimes improve model accuracy, especially if the outliers are due to errors or noise in the data. However, it can also reduce accuracy if the outliers represent genuine, important patterns in the data. It is crucial to verify the reason for the presence of outliers. Without such an analysis, there is no justification for the removal of outliers.

In my case, after removing outliers using Z-score, the accuracy of your models decreased to around 77-79%. This suggests that the outliers I removed might have contained valuable information for the model to learn from.

````python
# Get predicted probabilities for both algorithms
y_prob_lgb = best_model.predict_proba(x_val)[:, 1]
y_prob_xgb = best_model_xgb.predict_proba(x_val)[:, 1]
````

````python
# Calculate ROC curve and AUC for LightGBM
fpr_lgb, tpr_lgb, _ = roc_curve(y_val, y_prob_lgb)
roc_auc_lgb = auc(fpr_lgb, tpr_lgb)

# Calculate ROC curve and AUC for XGBoost
fpr_xgb, tpr_xgb, _ = roc_curve(y_val, y_prob_xgb)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)
````

````python
# Plot ROC curves
plt.figure()
plt.plot(fpr_lgb, tpr_lgb, color='darkorange', lw=2, label='LightGBM (AUC = %0.2f)' % roc_auc_lgb)
plt.plot(fpr_xgb, tpr_xgb, color='navy', lw=2, label='XGBoost (AUC = %0.2f)' % roc_auc_xgb)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
````

The **ROC (Receiver Operating Characteristic) curve** evaluates the performance of a classification model by plotting the **True Positive Rate (Recall)** against the **False Positive Rate** at various classification thresholds.

### **Key Observations from Your ROC Curve**:
1. **Comparison of Models**:
   - The **XGBoost model (AUC = 0.88)** performs slightly better than the **LightGBM model (AUC = 0.87)** in distinguishing between the two classes.
   - A higher **AUC (Area Under the Curve)** indicates a better ability to discriminate between positive and negative classes.

2. **True Positive Rate vs. False Positive Rate**:
   - A model with a **higher curve** (closer to the top-left) is generally better because it achieves a high true positive rate with a lower false positive rate.
   - Both models perform well since their curves are far from the diagonal (random classifier).

3. **Threshold Selection**:
   - The ROC curve helps in selecting an optimal threshold for classification.
   - If false positives are more costly, you might choose a threshold that reduces the false positive rate, even if it slightly lowers recall.

### **Final Conclusion**:
- Both **XGBoost and LightGBM** models are performing well.
- **XGBoost has a slight edge in AUC (0.88 vs. 0.87)**, meaning it is marginally better at distinguishing between classes.
- If precision (avoiding false positives) or recall (avoiding false negatives) is more important, you can fine-tune the decision threshold based on the ROC curve.

# Detailed Report: Comparative Analysis of XGBoost and LightGBM on Titanic Dataset

In this report, I present a detailed summary of my comparative analysis of two gradient boosting models—XGBoost and LightGBM—applied to the Titanic dataset. I utilized the training data (`Titanic_train.csv`) and test data (`Titanic_test.csv`) that I loaded into my notebook to predict passenger survival. My analysis focuses on the ROC (Receiver Operating Characteristic) curve results provided, and I’ve expanded on the comparative performance and practical implications based on the dataset’s features and modeling outcomes.

## Dataset Overview

Before diving into the results, I explored the Titanic training dataset, which contains 891 rows and 12 columns: `PassengerId`, `Survived`, `Pclass`, `Name`, `Sex`, `Age`, `SibSp`, `Parch`, `Ticket`, `Fare`, `Cabin`, and `Embarked`. The target variable, `Survived`, is binary (0 for not survived, 1 for survived), making this a classification task. Features like `Pclass` (passenger class), `Sex`, `Age`, and `Fare` likely play significant roles in survival prediction, though I noticed missing values in `Age` and `Cabin` that might require preprocessing (e.g., imputation or exclusion) before modeling.

## Comparative Analysis Results

I evaluated the performance of XGBoost and LightGBM using the ROC curve, which plots the True Positive Rate (TPR, or recall) against the False Positive Rate (FPR) across various thresholds. Below are my detailed observations:

### 1. Model Performance Comparison
- **XGBoost**: I found that XGBoost achieved an AUC (Area Under the Curve) of 0.88. This high AUC suggests that XGBoost effectively distinguishes between passengers who survived and those who didn’t, leveraging features like `Sex` (where females likely had higher survival rates) and `Pclass` (where first-class passengers might have had priority).
- **LightGBM**: In contrast, LightGBM yielded an AUC of 0.87, just 0.01 below XGBoost. While still impressive, this indicates a slightly lower ability to separate the classes. LightGBM’s performance might stem from its optimization for speed, potentially at a minor cost to accuracy in this case.
- **Interpretation**: An AUC closer to 1 signifies better discriminative power, and with XGBoost at 0.88 versus LightGBM at 0.87, I conclude that XGBoost has a slight edge. However, both models far exceed the baseline of 0.5 (random guessing), indicating strong predictive capabilities.

### 2. ROC Curve Analysis: TPR vs. FPR
- When I examined the ROC curves, I noticed that both models’ curves rise sharply toward the top-left corner, a sign of robust performance. The top-left corner represents an ideal scenario: high TPR (correctly identifying survivors) with low FPR (incorrectly predicting survival for non-survivors).
- XGBoost’s curve sits slightly higher than LightGBM’s, suggesting it achieves a better balance at most thresholds. For example, at a given FPR of 0.2, XGBoost might yield a TPR of 0.85, while LightGBM might reach 0.83 (hypothetical values based on AUC proximity).
- Both curves being well above the diagonal line (AUC = 0.5) reassure me that neither model is merely guessing; they’re effectively learning patterns from the data, such as the influence of `Sex` or `Fare`.

### 3. Threshold Selection Insights
