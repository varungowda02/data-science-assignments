```python
feature_df = pd.get_dummies(df[['genre', 'type']], columns=['genre', 'type'], prefix=['genre', 'type'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
feature_df['rating_scaled'] = scaler.fit_transform(df[['rating']])
```

```python
final_df = pd.concat([feature_df, df[['name']]], axis=1)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(anime_name, final_df, top_n=10):
  """Recommends anime based on cosine similarity.

  Args:
    anime_name: The name of the anime to find recommendations for.
    final_df: The dataframe containing the features for all anime.
    top_n: The number of recommendations to return.

  Returns:
    A list of recommended anime names.
  """

  # Get the index of the input anime
  anime_index = final_df[final_df['name'] == anime_name].index[0]

  # Calculate cosine similarity between the input anime and all other anime
  cosine_similarities = cosine_similarity(final_df.drop('name', axis=1), final_df.drop('name', axis=1))

  # Get the similarity scores for the input anime
  similarity_scores = list(enumerate(cosine_similarities[anime_index]))

  # Sort the anime by similarity score in descending order
  similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

  # Get the top N most similar anime (excluding the input anime itself)
  top_anime_indices = [i[0] for i in similarity_scores[1:top_n+1]]

  # Return the names of the recommended anime
  recommended_anime = final_df['name'].iloc[top_anime_indices].tolist()

  return recommended_anime
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(anime_name, final_df, top_n=10, threshold=0.8):
  """Recommends anime based on cosine similarity with a threshold.

  Args:
    anime_name: The name of the anime to find recommendations for.
    final_df: The dataframe containing the features for all anime.
    top_n: The maximum number of recommendations to return.
    threshold: The minimum similarity score for an anime to be recommended.

  Returns:
    A list of recommended anime names, or an empty list if the anime is not found.
  """

  # Check if anime_name exists in final_df
  if anime_name not in final_df['name'].values:
    return []  # Return empty list if anime not found

  # Get the index of the input anime
  anime_index = final_df[final_df['name'] == anime_name].index[0]

  # Calculate cosine similarity between the input anime and all other anime
  cosine_similarities = cosine_similarity(final_df.drop('name', axis=1), final_df.drop('name', axis=1))

  # Get the similarity scores for the input anime
  similarity_scores = list(enumerate(cosine_similarities[anime_index]))

  # Filter anime based on threshold
  filtered_scores = [i for i in similarity_scores if i[1] >= threshold]

  # Sort the filtered anime by similarity score in descending order
  filtered_scores = sorted(filtered_scores, key=lambda x: x[1], reverse=True)

  # Get the top N most similar anime (excluding the input anime itself)
  top_anime_indices = [i[0] for i in filtered_scores[1:top_n+1]]

  # Return the names of the recommended anime
  recommended_anime = final_df['name'].iloc[top_anime_indices].tolist()

  return recommended_anime
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(final_df, test_size=0.2, random_state=42)
```

```python
def get_relevant_anime(anime_name, df):
    """
    Gets a list of relevant anime based on shared genres.

    Args:
        anime_name: The name of the target anime.
        df: The original dataframe containing anime data.

    Returns:
        A list of anime names considered relevant to the target anime.
    """

    # Get the genres of the target anime
    target_genres = df[df['name'] == anime_name]['genre'].iloc[0].split(', ')

    # Find other anime with at least one shared genre
    relevant_anime = df[df['genre'].str.contains('|'.join(target_genres), na=False)]['name'].tolist()

    # Remove the target anime itself from the list
    relevant_anime.remove(anime_name)

    return relevant_anime
```

```python
from sklearn.metrics import classification_report

# Evaluation
y_true = []  # True relevance labels
y_pred = []  # Predicted relevance labels

for target_anime in test_df['name'].head(50):
    top_n = 10  # Define top_n here
    recommendations = recommend_anime(target_anime, train_df, top_n=top_n)

    # Get relevant anime for the target anime
    relevant_anime = get_relevant_anime(target_anime, df)

    # Create true and predicted labels for each recommendation
    for anime in recommendations:
        y_true.append(1 if anime in relevant_anime else 0) # 1 if relevant, 0 otherwise
        y_pred.append(1)  # We predict all recommendations as relevant

    # If there were fewer than top_n recommendations, add 0s for true labels
    num_missing_recs = top_n - len(recommendations) # Use the defined top_n
    y_true.extend([0] * num_missing_recs)
    y_pred.extend([0] * num_missing_recs)


# Generate the classification report
print(classification_report(y_true, y_pred))
```
