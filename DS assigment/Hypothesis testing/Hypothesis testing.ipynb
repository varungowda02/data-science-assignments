#Hypothesis Testing

###Background:
Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.

###Objective:
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.

###Data Provided:
* The theoretical weekly operating cost model: W = $1,000 + $5X  
* Sample of 25 restaurants with a mean weekly cost of Rs. 3,050  
* Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


___________________________________________________________________________________________________________________________

##Assignment Tasks:

________________

###1. State the Hypothesis statement:

#### Given Information
The company's theoretical cost model is given by:  

$$
W = 1000 + 5X
$$

where \( X \) (units produced per week) follows a normal distribution with:  

- Mean $$(\mu_X) = 600$$
- Standard deviation $$(\sigma_X) = 25$$

Using this model, we calculate the theoretical mean weekly cos*:

$$
\mu_W = 1000 + 5(600)
$$

$$
\mu_W = 1000 + 3000 = 4000
$$

#### Hypothesis Statements
- Null Hypothesis $( H_0 )$: The actual mean weekly operational cost is equal to the company's theoretical cost model.

$$
H_0: \mu_W = 4000
$$

- Alternative Hypothesis $(H_A)$: The actual mean weekly operational cost is greater than the theoretical cost model, as claimed by the restaurant owners.  

$$
H_A: \mu_W > 4000
$$

This is a one-tailed (right-tailed) hypothesis test, as we are testing whether the observed mean cost is significantly greater than 4000.


_____________________

### 2. Calculate the Test Statistic:

Use the following formula to calculate the test statistic (t):
where:  
* $\bar{x}$
 = sample mean weekly cost (Rs. 3,050)  
* μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)  
* σ = 5 * 25 units  
* n = sample size (25 restaurants)  

```python
# Import necessary libraries
import numpy as np
import scipy.stats as stats  # Statistical functions
```

```python
# Given data
sample_mean = 3050  # Mean weekly cost from the sample of 25 restaurants
theoritical_mean = 1000 + 5 * (600)  # Theoretical mean weekly cost from the cost model (W = 1000 + 5X)
std_dev = 5 * 25  # Standard deviation of the weekly cost (5 * σ_X)
n = 25  # Sample size (number of restaurants)
```

```python
# Calculate the test statistic using the t-formula:
# t = (sample_mean - theoretical_mean) / (standard deviation / sqrt(sample size))
t_statistic = (sample_mean - theoritical_mean) / (std_dev / np.sqrt(n))
t_statistic
```

If the t_statistic is large enough (i.e., greater than the critical value), it suggests that the difference is statistically significant.

______________________________

###3. Determine the Critical Value:
Using the alpha level of 5% (α = 0.05), determine the critical value from the standard normal (Z) distribution table.


```python
# Step 3: Determine the critical value from the standard normal Z-distribution
alpha = 0.05  # Significance level (5%)
z_critical = stats.norm.ppf(1 - alpha)  # Critical value from the Z-table for a one-tailed test
z_critical
```

If our t_statistic is greater than the critical value, it means the observed sample mean is significantly different from 4000, and we reject H₀.

______________________________________________

###4. Make a Decision:
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.



```python
# Step 4: Compare the test statistic with the critical value
if t_statistic > z_critical:
    print("Decision: Reject Null Hypothesis")
else:
    print("Decision: Fail to Reject Null Hypothesis")
```

###5. Conclusion:
Based on the decision in step 4, conclude whether there is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.

We fail to reject the null hypothesis as the data does not provide sufficient evidence to support the claim that costs have increased

_____

### Final Conclusion Based on the Analysis**

After conducting the hypothesis test, we found that the test statistic (t_statistic) is significantly lower than the critical value (z-score at α = 0.05). This means that the sample mean weekly cost (3050) is not statistically greater than the theoretical mean cost (4000) at a 5% significance level.  

Since t_statistic < z_critical, we fail to reject the null hypothesis (H₀).  

Thus, there is not enough statistical evidence to support the restaurant owners' claim that the weekly operational costs have increased compared to the company's theoretical cost model.  

While restaurant owners perceive an increase in costs, our statistical test does not confirm this claim based on the given data. Further investigation or a different approach may be needed to determine if other factors are influencing operational costs.
