### 1. **Normalization & Standardization**  
Normalization and standardization are techniques used to scale numerical data to improve the performance of machine learning algorithms.  

#### **Normalization**  
Normalization rescales the data to a fixed range, typically \([0,1]\) or \([-1,1]\). This ensures that all features contribute equally to the model.  
The **Min-Max Scaling** formula is:  

$$
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
$$  

where:  
- $X$ is the original value  
- $X_{min}$ and $X_{max}$ are the minimum and maximum values in the feature  

**Use case:** Normalization is preferred when the data does not follow a Gaussian (normal) distribution or when using models like KNN and Neural Networks.  

#### **Standardization**  
Standardization transforms the data to have a mean of 0 and a standard deviation of 1. It maintains the original distribution of the data.  
The **Z-score Scaling** formula is:  

$$
X' = \frac{X - \mu}{\sigma}
$$  

where:  
- $\mu$ is the mean of the feature  
- $\sigma$ is the standard deviation  

**Use case:** Standardization is used when data follows a normal distribution or when applying models like logistic regression, SVM, and PCA.

### 2. **Techniques to Address Multicollinearity in Multiple Linear Regression**  
Multicollinearity occurs when independent variables are highly correlated, affecting model interpretability.  

#### **Methods to Handle Multicollinearity:**  

1. **Variance Inflation Factor (VIF):**  
   - VIF quantifies the level of multicollinearity by measuring how much the variance of a regression coefficient is inflated.  
   - Formula:  

   $$
   VIF = \frac{1}{1 - R^2}
   $$  

   - If \(VIF > 10\), multicollinearity is high, and the variable should be removed or transformed.  

2. **Removing Highly Correlated Features:**  
   - Compute the correlation matrix and drop one of the highly correlated features.  

3. **Principal Component Analysis (PCA):**  
   - PCA transforms correlated variables into a set of uncorrelated principal components.  

4. **Regularization (Ridge or Lasso Regression):**  
   - **Ridge Regression:** Uses L2 regularization, which shrinks coefficients but keeps all features.  
   - **Lasso Regression:** Uses L1 regularization, which can eliminate some coefficients, making the model sparse.  
   - Ridge regression formula:  

     $$
     \beta = (X^T X + \lambda I)^{-1} X^T Y
     $$  

     where $\lambda$ is the regularization parameter.  

These techniques help improve model stability, interpretation, and predictive performance.


### **Regularization Techniques: Ridge & Lasso Regression**  
Regularization methods introduce a penalty to the regression model, preventing overfitting and reducing the impact of multicollinearity.  

---

## **Ridge Regression (L2 Regularization)**  
Ridge regression adds an L2 penalty, which shrinks the regression coefficients but does not eliminate any features.  
The cost function is modified as:  

$$
J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2
$$  

where:  
- $J(\beta)$ is the cost function  
- $y_i$ is the actual value  
- $\hat{y}_i$ is the predicted value  
- $\beta_j$ are the regression coefficients  
- $\lambda$ is the regularization parameter  

### **Impact of Ridge Regression:**  
- Reduces the effect of correlated variables without eliminating them.  
- Works well when all features contribute to the prediction.  

---

## **Lasso Regression (L1 Regularization)**  
Lasso regression adds an L1 penalty, which can shrink some coefficients to zero, effectively performing feature selection.  
The cost function is:  

$$
J(\beta) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |\beta_j|
$$  

### **Impact of Lasso Regression:**  
- Removes irrelevant features by setting some coefficients to zero.  
- Helps in feature selection while addressing multicollinearity.  

---

## **Choosing Between Ridge & Lasso:**  
- **Use Ridge** if all features are important but highly correlated.  
- **Use Lasso** if feature selection is needed to reduce complexity.  
- **Elastic Net** is a combination of both Ridge and Lasso for better performance.
