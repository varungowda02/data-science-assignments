```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for i in range(2, 11):  # Silhouette score is not defined for k=1
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(df)
    silhouette_scores.append(silhouette_score(df, labels))

# Plot the silhouette scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()

# Interpretation:
# Choose the k value that maximizes the silhouette score.
# Higher silhouette scores indicate better-defined clusters.
```

```python
# Calculate the average distance between each point and its k nearest neighbors
neighbors = NearestNeighbors(n_neighbors=5)  # You can adjust the number of neighbors (k)
neighbors_fit = neighbors.fit(df)
distances, indices = neighbors_fit.kneighbors(df)

# Sort and plot the distances
distances = np.sort(distances, axis=0)
distances = distances[:, 1]  # Get the distances to the kth nearest neighbor
plt.plot(distances)
plt.title('Elbow Method for Determining eps')
plt.xlabel('Data Points (Sorted by Distance)')
plt.ylabel('Distance to kth Nearest Neighbor')
plt.show()
```

```python
# Define linkage criteria to test
linkage_criteria = ['ward', 'complete', 'average', 'single']

# Iterate through linkage criteria and evaluate clustering performance
for linkage in linkage_criteria:
    # Perform hierarchical clustering
    hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage=linkage)
    hc_labels = hc.fit_predict(df)

    # Calculate silhouette score
    silhouette = silhouette_score(df, hc_labels)

    # Print results
    print(f"Linkage: {linkage}, Silhouette Score: {silhouette}")

# Choose the linkage criterion with the highest silhouette score
best_linkage = linkage_criteria[np.argmax([silhouette_score(df, AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage=linkage).fit_predict(df)) for linkage in linkage_criteria])]
print(f"\nBest Linkage Criterion: {best_linkage}")
```

```python
from sklearn.cluster import KMeans

# Define range of K values to test
k_values = range(1, 11)

# Calculate WCSS for each K value
wcss = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()

# Choose the K value at the "elbow" point
# This might be slightly inaccurate, but try the following:
# visually inspect the elbow curve to identify the "elbow" point which corresponds to the optimal K value
# and update the variable 'optimal_k' with the chosen value.
optimal_k = 3
```

```python
# Perform K-means clustering with the optimal K value
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
kmeans_labels = kmeans.fit_predict(df)

# Calculate silhouette score
silhouette = silhouette_score(df, kmeans_labels)

# Print results
print(f"Optimal k: {optimal_k}, Silhouette Score: {silhouette}")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

# Define ranges for epsilon and minPts
eps_values = np.arange(0.1, 2.0, 0.1)
min_samples_values = range(2, 10)

# Store results for all combinations
results = []

# Iterate through parameter combinations and evaluate clustering performance
for eps in eps_values:
    for min_samples in min_samples_values:
        # Perform DBSCAN clustering
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan_labels = dbscan.fit_predict(df)

        # Calculate silhouette score (if there are at least 2 clusters)
        if len(np.unique(dbscan_labels)) > 1:
            silhouette = silhouette_score(df, dbscan_labels)
            results.append((eps, min_samples, silhouette))
            print(f"eps: {eps}, minPts: {min_samples}, Silhouette Score: {silhouette}")

# Find the best parameters
best_eps, best_min_samples, best_silhouette = max(results, key=lambda x: x[2])

# Print the best parameters
print(f"\nBest eps: {best_eps}, Best minPts: {best_min_samples}, Silhouette Score: {best_silhouette}")
```
