```python
from sklearn.metrics import roc_curve, auc
```

```python
# Get unique classes from y_test
classes = sorted(y_test.unique())
n_classes = len(classes)

# Calculate number of rows and columns for subplots
n_cols = 2  # Two columns
n_rows = (n_classes + 1) // n_cols  # Calculate rows, ensuring enough space
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10, 5 * n_rows))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Iterate through each class and plot on subplots
for i, class_to_analyze in enumerate(classes):
    # Create a binary target variable for the current class
    y_test_binary = [1 if y == class_to_analyze else 0 for y in y_test]

    # Get predicted probabilities for the current class
    y_prob = model.predict_proba(x_test)[:, class_to_analyze]

    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_test_binary, y_prob)

    # Calculate ROC AUC
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve on the current subplot
    axes[i].plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    axes[i].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    axes[i].set_xlim([0.0, 1.0])
    axes[i].set_ylim([0.0, 1.05])
    axes[i].set_xlabel('False Positive Rate')
    axes[i].set_ylabel('True Positive Rate')
    axes[i].set_title(f'ROC for Class {class_to_analyze}')
    axes[i].legend(loc="lower right")

# Hide any extra subplots if the number of classes is odd
for j in range(n_classes, n_rows * n_cols):
    axes[j].axis('off')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

The decision tree visualization provides insights into how the model classifies data based on different features and their importance in splitting the dataset. Below is an analysis of the tree structure:

### **Key Observations:**
1. **Root Node (Topmost Split) - `cp (chest pain type)`**
   - The first split is based on `cp <= 0.5`, indicating that chest pain is the most important factor in classification.
   - The **Gini index** at this level is **0.698**, meaning the data is quite mixed in terms of class distribution.
   - The **samples = 726**, showing the entire dataset is split at this node.
   - The **value** list represents the count of samples in each class.

2. **Left Branch (cp <= 0.5) - `age`**
   - The next split is based on **age**, suggesting that younger and older individuals show different patterns in classification.
   - The Gini index **0.734** is still relatively high, indicating class impurity.
   - Further splits occur based on **exang (exercise-induced angina)** and **trestbps (resting blood pressure).**

3. **Right Branch (cp > 0.5) - `chol (cholesterol)`**
   - The next major factor influencing classification is **cholesterol levels.**
   - The Gini index here **(0.468)** is lower, meaning this branch has more homogeneous classes compared to the left branch.
   - Further splits occur based on **age and thalch (maximum heart rate achieved).**

### **Gini Index Interpretation:**
- **Lower Gini Index (closer to 0):** More homogeneity in classification (pure nodes).
- **Higher Gini Index (closer to 1):** More impurity, indicating mixed classifications.

### **Class Distribution:**
- The final leaf nodes show **pure and mixed class distributions.**
- Some nodes have **Gini = 0**, meaning all samples in that node belong to the same class.
- Others have **higher Gini values**, showing mixed classification.

### **Conclusion:**
- The decision tree primarily **relies on chest pain type (`cp`), age, cholesterol, and exercise-induced angina (`exang`)** for classification.
- **Lower Gini nodes indicate stronger predictive splits, while higher Gini nodes show overlapping data points.**
- Feature selection and pruning could help simplify the tree and avoid overfitting.

### **Interview Questions**
#### **1. What are common hyperparameters of Decision Tree models?**
- **max_depth**: Limits the depth of the tree to prevent overfitting.
- **min_samples_split**: Minimum samples required to split a node.
- **criterion**: Splitting strategy (`gini` or `entropy`).

#### **2. Difference between Label Encoding and One-Hot Encoding**
- **Label Encoding**: Assigns unique integer values to categories (e.g., Male = 1, Female = 0). Used when categories have an order.
- **One-Hot Encoding**: Creates separate binary columns for each category. Used when categories have no order.
