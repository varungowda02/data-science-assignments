```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Function to create the model
def create_model(neurons=64, activation='relu'):
    model = Sequential([
        Input(shape=(x_train.shape[1],)),  # Define input shape using Input layer
        Dense(neurons, activation=activation),
        Dense(32, activation=activation),
        Dense(len(y.unique()), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Example usage
model = create_model(neurons=128, activation='relu')
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
```

```python
best_accuracy = 0
best_params = {}

for neurons in [32, 64, 128]:
    for activation in ['relu', 'tanh']:
        model = Sequential([
            Dense(neurons, activation=activation, input_shape=(x_train.shape[1],)),
            Dense(32, activation=activation),
            Dense(len(y.unique()), activation='softmax')
        ])
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        model.fit(x_train, y_train, epochs=10, batch_size=32, verbose=0)

        accuracy = model.evaluate(x_test, y_test, verbose=0)[1]  # Get accuracy
        print(f"Neurons: {neurons}, Activation: {activation}, Accuracy: {accuracy}")

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'neurons': neurons, 'activation': activation}

print(f"Best Hyperparameters: {best_params}, Best Accuracy: {best_accuracy}")
```

### Analysis of Output

The given code performs hyperparameter tuning for an Artificial Neural Network (ANN) classification model using different numbers of neurons (32, 64, 128) and activation functions (`relu`, `tanh`). ### Variability in Model Performance Across Runs  

When training deep learning models, it is common to observe slight variations in performance metrics (such as accuracy, precision, recall, and F1-score) across different runs, even without changing the hyperparameters. In this case, the model initially achieved **0.92 accuracy** in one run and **0.91 accuracy** in a subsequent run without restarting the kernel. This variability can be attributed to several factors:  

1. **Random Initialization of Weights:**  
   - Neural networks start with randomly initialized weights.  
   - Different initial weights can lead to different optimization paths, slightly affecting the final accuracy.  

2. **Stochastic Nature of Training:**  
   - The training process involves **stochastic gradient descent (SGD) or Adam optimizer**, both of which introduce randomness.  
   - The way mini-batches are sampled during training can also cause slight variations in learning.  

3. **Floating-Point Precision:**  
   - Deep learning computations use floating-point arithmetic, which can introduce small numerical variations in results across runs.  

4. **Non-Deterministic Hardware Operations:**  
   - Certain operations, especially those involving GPU acceleration, are non-deterministic due to parallel processing, leading to minor differences in outcomes.  

### Mitigating Variability  
To ensure more consistent results across runs, the following techniques can be used:  
- **Setting a Random Seed:** Using `numpy`, `tensorflow`, or `random` seed settings can help reduce variability.  
- **Using Deterministic Operations:** Disabling certain non-deterministic GPU operations can help stabilize results.  
- **Running Multiple Trials and Averaging Results:** Instead of relying on a single run, averaging performance over multiple runs provides a more robust evaluation.  

### Conclusion  
Variations in accuracy across runs are normal due to the stochastic nature of deep learning training. These fluctuations do not indicate issues with the model but rather reflect the inherent randomness in neural network optimization.
